```python
import numpy as np
import matplotlib.pyplot as plt

def find_digit_index(digit, labels):
    return np.where(labels == digit)[0][0]

datasets = [('/content/drive/MyDrive/mnist_mix_all/ARDIS_train_test.npz', 'ARDIS'),
            ('/content/drive/MyDrive/mnist_mix_all/BanglaLekha_train_test.npz', 'BanglaLekha-Isolated'),('/content/drive/MyDrive/mnist_mix_all/Bangla_train_test.npz', 'CMATERdb 3.1.1'),
            ('/content/drive/MyDrive/mnist_mix_all/Devanagari_train_test.npz', 'CMATERdb 3.2.1'),('/content/drive/MyDrive/mnist_mix_all/Arabic_train_test.npz', 'CMATERdb 3.3.1'),
             ('/content/drive/MyDrive/mnist_mix_all/Telugu_train_test.npz', 'CMATERdb 3.4.1'),('/content/drive/MyDrive/mnist_mix_all/EMNIST_train_test.npz', 'EMNIST'),
            ('/content/drive/MyDrive/mnist_mix_all/Farsi_train_test.npz', 'Farsi'),('/content/drive/MyDrive/mnist_mix_all/ISI_Bangla_train_test.npz', 'ISI Bangla'),
             ('/content/drive/MyDrive/mnist_mix_all/Kannada_train_test.npz', 'Kannada-mnist'),('/content/drive/MyDrive/mnist_mix_all/MNIST_MIX_train_test.npz', 'MADBase'),
             ('/content/drive/MyDrive/mnist_mix_all/Tibetan_train_test.npz', 'TibetanMNIST'),('/content/drive/MyDrive/mnist_mix_all/Urdu_train_test.npz', 'Urdu'),
         ]

for dataset_path, language in datasets:
    data = np.load(dataset_path)
    X_train, y_train = data['X_train'], data['y_train']

    # Plot one example of each digit
    fig, axes = plt.subplots(1, 10, figsize=(15, 3))

    # Set the title to the left of the images
    fig.suptitle(f"{language}", x=0.05, y=0.95, ha='left', fontsize=14)

    for i in range(10):
        digit_index = find_digit_index(i, y_train)
        axes[i].imshow(X_train[digit_index], cmap='gray')
        axes[i].set_title(str(i))
        axes[i].axis('off')

    plt.subplots_adjust(left=0.05, right=0.95, top=1, bottom=0.2, wspace=0.5)
    plt.show()
```
